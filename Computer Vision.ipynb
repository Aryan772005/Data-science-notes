```python
image = cv2.imread("image1.jpg")
```

```python
image_rgb = cv2.cvtColor(image , cv2.COLOR_BGR2RGB)
```

```python
height , width = image.shape[:2]
```

```python
center = (width // 2, height//2)
```

```python
# roatation matrix ko kaise ayega
rotation_matrix = cv2.getRotationMatrix2D(center, 60, 1)

# width , height ko adjust
rotated = cv2.warpAffine(image, rotation_matrix, (width, height))

# Convert to RGB for correct colors in Matplotlib
rotated_rgb = cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB)

# Display the final rotated image

plt.imshow(rotated_rgb)
plt.title("hogya(60Degreesplt.title ka use sey)")
plt.axis('off')
plt.show()
```

```python


# 1. Load the original image
image = cv2.imread("image1.jpg")

# 2. Convert to RGB for correct colors in Matplotlib
original_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# 3. Convert to Grayscale
original_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)


# --- PRINT PHOTO 1: ORIGINAL COLOR ---
plt.figure() # Creates a clean, separate frame
plt.imshow(original_rgb)
plt.title("Original Color Image")
plt.axis('off')
plt.show() # This prints the first image and closes the block


# --- PRINT PHOTO 2: GRAYSCALE ---
plt.figure() # Creates another clean, separate frame
plt.imshow(original_gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')
plt.show() # This prints the second image right below it
```

```python


# 1. Create a blank dark canvas
canvas_rect = np.zeros((400, 600, 3), dtype=np.uint8) + 50

# 2. Draw a RECTANGLE
# Coordinates: (start_x, start_y), (end_x, end_y)
start_point = (100, 80)
end_point = (400, 300)
color_bgr = (0, 0, 255) # Red in BGR
thickness = 3

cv2.rectangle(canvas_rect, start_point, end_point, color_bgr, thickness)

# 3. Convert to RGB and display
canvas_rect_rgb = cv2.cvtColor(canvas_rect, cv2.COLOR_BGR2RGB)
plt.imshow(canvas_rect_rgb)
plt.title("ye rectangle")
plt.axis('off')
plt.show()
```
