```python
password = "redhat"
```

```python
def check_password():
    user_password=input("Enter your pass:")
    if user_password == password:
                return True
```

```python
def write_note():
    if check_password():
        note = input("Enter Secret note: ")
        with open("data.txt.txt", "w") as file:
            file.write(note)
```

```python
def read_note():
    if check_password():
        file = open("data.txt.txt", "r")
        data = file.read()
        file.close()
        print("\n your secrete note is :")
        print(data)
```

```python
def append_note():
    if check_password():
        note = input("Enter note: ")

        file = open("data.txt.txt", "a")
        file.write("\n" + note)
        file.close()

        print("Note Added Successfully")
```

```python
def menu():
    while True:
        
        print("""
========== Secret Note Key ==========
1. Write Note
2. Read Note
3. Append Note
4. Exit
""")

        choice = input("Enter your choice: ")

        if choice == "1":
            write_note()
        elif choice == "2":
            read_note()
        elif choice == "3":
            append_note()
        elif choice == "4":
            print("Goodbye!")
            break
        else:
            print("Invalid choice. Try again.")
```

```python
menu()
```
