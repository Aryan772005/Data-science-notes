```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
```

isme mai label encoder fumction sey sab strings ko 0 and 1 me convert kardiya

```python
categorial = ["Sex","Ethnicity","Physical_Activity_Level","Alcohol_Consumption","Smoking_Status"]
```

```python
for col in categorial:
    df[col] = encoder.fit_transform(df[col])
    
```

```python
df.head()
```

```python
df = df.drop(columns=["Alcohol_Co =nsumption"])
```

```python
Y = df["Family_History_of_Diabetes"]
```

```python
X = df.drop("Family_History_of_Diabetes",axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split( 
    X,
    Y,
    test_size = 0.2,
    random_state=42
    
)
```

```python
X_train
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(random_state=42)
```

```python
model.fit(X_train,Y_train)
```

```python
Y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix , classification_report
```

```python
accuracy_score(Y_test,Y_pred)
```

```python
cm = confusion_matrix(Y_test,Y_pred)
```

```python
cm
```

```python
sns.heatmap(cm,
            annot=True,
            cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

Acurracy score check karne ke liye

```python
print(classification_report(Y_test, Y_pred))
```

```python
important = pd.DataFrame ({
    "Features" : X.columns,
    "Important" : model.feature_importances_
})
```

```python
important.sort_values(
    by="Important",
    ascending=False
)
```

```python

plt.figure(figsize=(12, 8))


sns.barplot(
    x="Important", 
    y="Features", 
    data=important.sort_values(by="Important", ascending=False),
    palette="viridis"
)


plt.title("Feature Importance for Diabetes Prediction Model", fontsize=16, pad=15)
plt.xlabel("Importance Score", fontsize=12)
plt.ylabel("Features", fontsize=12)


plt.tight_layout()
plt.show()
```
