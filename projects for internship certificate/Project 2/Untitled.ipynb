```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Sample - Superstore.csv", encoding="latin1")
```

```python
df.head()
```

```python
print(df.columns)
```

```python
print("Shape of Dataset:", df.shape)
```

```python
print(df.dtypes)
```

```python
df.isnull().sum()
```

```python
print("Before:", df.shape)

df.drop_duplicates(inplace=True)

print("After:", df.shape)
```

```python
df.describe()
```

```python


df["Profit_Status"] = np.where(
    df["Profit"] > 0,
    "Profit",
    "Loss"
)

df[["Profit","Profit_Status"]].head()
```

```python


df["High_Value_Order"] = np.where(
    df["Sales"] > 1000,
    "Yes",
    "No"
)

df[["Sales","High_Value_Order"]].head()
```

```python
def discount_category(discount):

    if discount <= 0.10:
        return "Low"

    elif discount <= 0.20:
        return "Medium"

    else:
        return "High"


df["Discount_Category"] = df["Discount"].apply(discount_category)

df[["Discount","Discount_Category"]].head()
```

```python
def sales_category(sales):

    if sales < 100:
        return "Low Sales"

    elif sales < 500:
        return "Medium Sales"

    else:
        return "High Sales"


df["Sales_Category"] = df["Sales"].apply(sales_category)

df[["Sales","Sales_Category"]].head()
```

```python
total_sales = df["Sales"].sum()

print("Total Sales =", round(total_sales,2))
```

```python
total_profit = df["Profit"].sum()

print("Total Profit =", round(total_profit,2))
```

```python
average_sales = df["Sales"].mean()

print("Average Sales =", round(average_sales,2))
```

```python
category_sales = df.groupby("Category")["Sales"].sum()

print(category_sales)
```

```python
top_states = df.groupby("State")["Sales"].sum().sort_values(ascending=False)

print(top_states.head())
```

```python
top_customers = df.groupby("Customer Name")["Sales"].sum().sort_values(ascending=False)

print(top_customers.head())
```

```python
print(df["Profit_Status"].value_counts())
```

```python
print(df["High_Value_Order"].value_counts())
```

```python
print("Total Records :", len(df))

print("Total Sales :", round(df["Sales"].sum(),2))

print("Total Profit :", round(df["Profit"].sum(),2))

print("Average Sales :", round(df["Sales"].mean(),2))

print("Profit Orders :", (df["Profit_Status"]=="Profit").sum())

print("Loss Orders :", (df["Profit_Status"]=="Loss").sum())

print("High Value Orders :", (df["High_Value_Order"]=="Yes").sum())
```

```python
df.to_csv("Processed_Superstore.csv", index=False)

print("Processed dataset saved successfully.")
```
